### 1.2 Estrutura dos dados

```python
import pandas as pd

df_matriculas = pd.DataFrame()
df_turmas = pd.DataFrame()

for i in range(2009,2019):
    
    for j in range(1,3):
        print(str(i)+'.'+str(j))
        df_matricula = pd.read_csv('matricula-de-'+str(i)+str(j)+'.csv',sep=';')
        df_matriculas = pd.concat([df_matriculas, df_matricula])
        
        df_turma = pd.read_csv('turmas-'+str(i)+str(j)+'.csv',sep=';')
        df_turmas = pd.concat([df_turmas, df_turma])

df_componentes = pd.read_csv('componentes-curriculares-presenciais.csv',sep=';')
        
# print("\nDataFrame das matrículas de 2015.1:", df_matriculas.shape)
# display(df_matriculas.head(3))

# print("\nDataFrame das turmas de 2015.1:", df_turmas.shape)
# display(df_turmas.head(3))

# print("\nDataFrame dos componentes curriculares:", df_componentes.shape)
# display(df_componentes.head(3))
```

O primeiro dataset representa a relação das matrículas em componentes dos cursos da instituição. Cada coluna significa:

- **descrição**: Situação final do discente nessa turma.
- **discente**: Identificador do discente da matrícula.
- **faltas_unidade**: Quantidade de faltas na unidade.
- **id_curso**: Identificador do curso do discente.
- **id_turma**: Identificador da turma do discente.
- **media_final**: Média final do discente.
- **nota**: Nota do discente na unidade.
- **numero_total_faltas**: Quantidade total de faltas.
- **reposição**: Indica se o discente ficou em reposição.
- **unidade**: Unidade corrente do período letivo.

O segundo dataset possui a relação de turmas dos cursos de nível médio, técnio, graduação e pós-graduação da UFRN. Suas colunas são:

- **id_turma**: Identificador da turma.
- **codigo_turma**: Código de identificação da turma.
- **siape**: Siape do docente interno.
- **matricula_docente_externo**: Matrícula do docente externo.
- **observação**: Observação para a turma.
- **id_componente_curricular**: Identificador do componente curricular da turma.
- **ch_dedicada_periodo**: Carga horária dedicada pelo docente à turma.
- **nível**: Nível de ensino da turma.
- **campus_turma**: Campus onde a turma é ofertada.
- **local**: Local físico de oferta da turma.
- **ano**: Ano de oferta da turma.
- **período**: Período de oferta da turma.
- **data_início**: Data de início da turma.
- **data_fim**: Data de término da turma.
- **descrição_horário**: Descrição textual da expressão de horário.
- **total_solicitações**: Total de solicitações de matrícula cadastradas para a turma.
- **capacidade_aluno**: Capacidade de discentes na turma.
- **tipo**: Identificador para informar o tipo da Turma.
- **distância**: Condicional para informar se a turma é oferta na modalidade a distância.
- **data_consolidação**: Data de consolidação da turma.
- **agrupadora**: Condicional para informar se a turma funciona como agrupadora para subturmas.
- **id_turma_agrupadora**: Corresponde ao id_turma da agrupadora desta turma.
- **qtd_aulas_lançadas**: Quantitativo de aulas lançadas pelo docente.
- **situação_turma**: Situação da turma.
- **convênio**: Convênio de vínculo com a turma.
- **modalidade_participantes**: Modalidade de educação para alunos destinados a turma.

O que importa do dataset de componentes:

- **id_componente**
- **nome**: nome da disciplina

### 1.3 Tabela resultante

#### 1.3.1 Filtrando as matriculas

```python
# filtrando os alunos de CeT
df_matriculas2 = df_matriculas[df_matriculas['id_curso']==10320810.0]
df_matriculas2.shape
```

```python
cols = ['descricao', 'discente', 'id_curso','id_turma', 'media_final']
df_matriculas2 = df_matriculas2[cols]
df_matriculas2.shape
```

#### 1.3.2 Merge com turmas

```python
df_turmas2 = df_turmas[(df_turmas['nivel_ensino'] == 'GRADUAÇÃO')]
cols_turma = ['id_turma', 'codigo_turma',
       'id_componente_curricular', 'ano', 'periodo',
       'data_inicio', 'data_fim', 'data_consolidacao',
       'agrupadora', 'id_turma_agrupadora']
df_turmas2 = df_turmas2[cols_turma]
df_turmas2.shape
```

```python
cols_componentes = ['id_componente', 'codigo', 'nome',
       'unidade_responsavel','pre_requisito', 'co_requisito',  
       'ano_programa', 'periodo_programa', 'curso_componente']
df_componentes2 = df_componentes[cols_componentes]
df_componentes2 = df_componentes2[df_componentes2['unidade_responsavel'] == 'ESCOLA DE CIÊNCIAS E TECNOLOGIA']
df_componentes2.shape
```

```python
dados = pd.merge(df_matriculas2, df_turmas2, on='id_turma', how='left')
dados = pd.merge(dados, df_componentes2, left_on='id_componente_curricular', right_on='id_componente', how='left')
dados.shape
```

```python
dados = dados.drop_duplicates()
dados.shape
```

#### 1.3.3 Filtragem dos alunos de 2015.1

```python
# conjunto de alunos que não serão analisados
series_Antes2015 = dados[dados['ano']<2015].discente

# selecionando apenas alunos que entraram a partir de 2015
df_depois2015 = dados[~dados['discente'].isin(series_Antes2015)]

df_depois2015['anoperiodo'] = (df_depois2015['ano'].map(str)+df_depois2015['periodo'].map(str))
df_depois2015.info()
```

```python
# conjunto de alunos que entraram em 2015.1
series_em20151 = df_depois2015[df_depois2015['anoperiodo']=='20151'].discente

# selecionando apenas alunos que entraram a partir de 2015
df_alunos20151 = df_depois2015[df_depois2015['discente'].isin(series_em20151)]
df_alunos20151.shape
```
