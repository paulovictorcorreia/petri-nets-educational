```python
import snakes.plugins
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
snakes.plugins.load(["gv", 'labels'], "snakes.nets", "nets")
from nets import *

path = 'images/'

net = PetriNet("mynet")
net.add_place(Place('p1', [dot]))
net.add_place(Place('p2'))
net.add_place(Place('p3', [dot]))
net.add_transition(Transition('t'))
net.add_input('p1', 't', Value(dot))
net.add_input('p3', 't', Value(dot))
net.add_output('p2', 't', Value(dot))
net.draw(f'{path}mynet33.png')
```

```python
t = net.transition('t')
m = t.modes()
t.fire(m.pop())
net.draw(f'{path}mynet2.ps')
```

## Preparing Petri Network Color

The next part of the work is to prepare the color of of marcations for the colored petri network system. We will use the students' characteristics described in the *students_final* dataframe, in which we dropped students with registration on the subject described as cancelled, indefered and locked. 

```python
import pandas as pd


class Student:
    def __init__(self, student_id):
        self.student_id = student_id
        self.subjects = []
    def __repr__(self):
        return str(self.student_id)
    def __str__(self):
        return f"Student id: {self.student_id}\nStudent's subjects history: {self.subjects}"
    def add_subject(self, discipline):
        self.subjects.append(discipline)
    def require_subject_info(self, students):
        '''
        Returns wether or not the student's subjects object is filled with the students data.
        If the subjects list was empty, it returns True and fills the list with all the necessary
        data of the specified student.
        Parameter:
        * students: students' dataframe with info of all students who entered S&T course in 2015.1.
        '''
        student_table = students[students['discente'] == self.student_id]
        rows_student = student_table.iterrows()
        if self.subjects == []:
            for row in rows_student:
                series = pd.Series(row)
                discipline = Discipline(
                    ident = row[1]['id_componente'],
                    name = row[1]['nome'],
                    semester = row[1]['anoperiodo'],
                    status = row[1]['descricao']
                )
                self.subjects.append(discipline)
            return True
        else:
            return False
        def correct_status(self):
            #corrigir a forma que representa o status naquela disciplina
            return True

        
class Discipline:
    def __init__(self, ident, name, semester, status):
        '''
        constructor of the class Discipline, it has 4 parameters:
        * ident: code of the discipline to identify on the components table
        * name: name of the subject
        * status: student's final status this subject
        * semester: semester in which the student attended to this classroom
        '''
        self.ident = ident
        self.name = name
        self.semester = semester
        self.status = status
    def __str__(self):
        return f'Component id: {self.ident}, \nDiscipline name: {self.name}, \nDiscipline semester: {self.semester}, \nStudent status: {self.status}\n'
    def __repr__(self):
        return f'Component id: {self.ident}, Discipline name: {self.name}, Discipline semester: {self.semester}, Student status: {self.status}\n'
        
teste = Student(student_id='6360ce81e659feee39452aa8cd111beb')
teste.require_subject_info(students_final)
print(teste)
```
